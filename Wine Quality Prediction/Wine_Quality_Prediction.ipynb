Importing the Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Data Collection

```python
# loading the dataset to a pandas dataframe
wine_dataset = pd.read_csv('/content/winequality-red.csv')
```

```python
# number of rows and columns
wine_dataset.shape
```

```python
# first 5 rows of the dataset
wine_dataset.head()
```

```python
# checking for missing values
wine_dataset.isna().sum()
```

```python
# statistical measures of the dataset
wine_dataset.describe()
```

```python
# number of values for each quality
sns.catplot(kind='count',hue='quality',x='quality', data=wine_dataset)
```

```python
# volatile acidity vs Quality
sns.barplot(x='quality', y='volatile acidity', data= wine_dataset, hue='quality')
```

```python
# citric acid vs Quality
sns.barplot(x='quality', y='citric acid', data= wine_dataset, hue='quality')
```

Correlation

1. Positive Correlation

2. Negative Correlation

```python
correlation = wine_dataset.corr()
```

```python
# constructing a heatmap to understand the correlation between the columns
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, annot=True, fmt='.1f', annot_kws={'size':10}, linecolor='white', cmap='coolwarm')
```

Data Preprocessing

```python
# separate the data and label
X = wine_dataset.drop('quality', axis=1)
print(X)
```

Label Binarization

```python
y = wine_dataset.quality.apply(lambda val: 1 if val>=6 else 0)
```

```python
print(y)
```

Train & Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, stratify=y, random_state=1)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

Model Training

Random Forest Classifier

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

Model Evaluation

Accuracy Score

```python
# checking accuracy score of model on training data
X_training_prediction = model.predict(X_train)
training_accuracy_score = accuracy_score(X_training_prediction, y_train)
```

```python
print('Accuracy score on training data: ', training_accuracy_score)
```

```python
# checking accuracy score of model on test data
X_test_prediction = model.predict(X_test)
test_accuracy_score = accuracy_score(X_test_prediction, y_test)
```

```python
print('Accuracy score on test data: ', test_accuracy_score)
```
